```python
df = pd.read_csv("archivo.csv")
df.columns = [
    "Time",
    "X",
    "State_ID",
    "Name",
    "Job",
    "Rank",
    "Action",
    "Location"
]

df.head(5)
```

```python
# Ensure datetime
df["Time"] = pd.to_datetime(df["Time"])

# Sort properly
df = df.sort_values(["State_ID", "Time"])

# Shift values inside each State_ID
df["Next_Action"] = df.groupby("State_ID")["Action"].shift(-1)
df["Next_Time"] = df.groupby("State_ID")["Time"].shift(-1)

# Keep valid Check In → Check Out pairs
sessions = df[
    (df["Action"] == "Check In") &
    (df["Next_Action"] == "Check Out")
].copy()

# Calculate duration
sessions["Duration"] = sessions["Next_Time"] - sessions["Time"]

# Aggregate total time per State_ID and Name
result = (
    sessions.groupby(["State_ID", "Name"])["Duration"]
    .sum()
    .reset_index()
)

# Convert to hours
result["Hours"] = result["Duration"].dt.total_seconds() / 3600

# Keep only desired columns
result = result[["State_ID", "Name", "Hours"]]

# Sort from least to most hours
result = result.sort_values("Hours").reset_index(drop=True)

# Optional: round hours
result["Hours"] = result["Hours"].round(2)

print(result)

result.to_csv("state_hours.csv", index=False, encoding="utf-8")

```
